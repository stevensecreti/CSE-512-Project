```python
df=pd.read_csv('US_youtube_trending_data.csv', usecols = ['video_id', 'publishedAt', 'trending_date', 'view_count', 'comment_count', 'categoryId'])
```

```python
print(df.values)
```

```python
def convert(date):
  temp=date.split('T')
  date=temp[0].split('-')
  tim=temp[1][:-1].split(':')
  ret=[]
  for d in date:
    ret.append(int(d))
  for t in tim:
    ret.append(int(t))
  return time.mktime(datetime.datetime(*ret).timetuple())
```

```python
data=df.values
ids=data[:, 0]
idx=np.argsort(ids)
data=data[idx]
```

```python
current_id=''
unique_videos=[]
for video in data:
  if video[0]==current_id:
    continue
  current_id=video[0]
  published_at=convert(video[1])
  trending_date=convert(video[3])
  unique_videos.append({'video_id': video[0], 'published_at':published_at, 'trending_date': trending_date, 'comment_view_ratio': video[4]/(video[5]+1), 'category': video[2]})
```

```python
print(unique_videos[0])
```

```python
with open('structure_to_category_epoch.json', 'w') as f:
  json.dump(unique_videos, f, indent=4)
```

extracting data for structure to likes/dislikes

```python
df=pd.read_csv('US_youtube_trending_data.csv', usecols = ['video_id', 'publishedAt', 'trending_date', 'view_count', 'comment_count', 'categoryId', 'likes', 'dislikes'])
```

```python
print(df.values[0])
```

```python
data=df.values
ids=data[:, 0]
idx=np.argsort(ids)
data=data[idx]
```

```python
current_id=''
unique_videos=[]
for video in data:
  if video[0]==current_id:
    continue
  current_id=video[0]
  published_at=convert(video[1])
  trending_date=convert(video[3])
  unique_videos.append({'video_id': video[0], 'published_at':published_at, 'trending_date':trending_date, 'view_count':video[4], 'comment_count': video[7], 'likes_to_dislikes_ratio': video[5]/(video[6]+1)})
```

```python
with open('structure_to_likes_to_dislikes_ratio_epoch.json', 'w') as f:
  json.dump(unique_videos, f, indent=4)
```
